#Algerian Forest Fires Dataset

## Data Set Information:

### The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

### 12 instances for each region.

### The period from June 2012 to September 2012. The dataset includes 11 attributes and 1 output attribute (class). The 244 instances have been classified into fire (138 classes) and not fire (106 classes) classes.

Attribute Information:

1. Date: (DD/MM/YYYY) Day, month (June to September), year (2012) data observations
2. Temp: temperature noon (temperature max) in Celsius degrees: 22 to 40
3. Relative Humidity in %: 21 to 90
4. Wind speed in km/h: 6 to 29
5. Fine Fuel Moisture Code (FFMC) index from the FWI system: 0.0 to 100
6. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
7. Drought Code (DC) index from the FWI system: 7.1 to 220.4
8. Fire Weather Index (FWI) index from the FWI system: 0.1 to 68
9. Initial Spread Index (ISI) index from the FWI system: 0.0 to 68
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire danger class: 0 to 5

Two classes, namely Fire and not Fire.

# Data Cleaning

```python
### Missing value
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1: "Bejaia Region Dataset"

2: "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"] = 1
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Remove the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
### Remove the 122nd row
df = df.drop(122).reset_index(drop=True)
```

```python
### Fix spaces in columns names   .strip() tm ra cc khong trng v cp nht n
df.columns = df.columns.str.strip()
df.columns
```

```python
df.head()
```

Changes the required columns as integer data

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

Changes the other columns to float data datatype

```python
objects = [features for features in df.columns if df[features].dtype == 'O']
```

```python
for i in objects:
  if i != 'Classes':
    df[i] = df[i].astype(float)
```
